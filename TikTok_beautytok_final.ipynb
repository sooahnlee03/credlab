```python
import json
import pandas as pd

"""
This function reads a JSON file and converts it to a pandas DataFrame.
"""
def read_json2csv(readin_file, readout_file):
    with open(readin_file, 'r') as file:
        readout_file = json.load(file)
        
    df = pd.json_normalize(readout_file["videos"])
    
    # format ids to be strings to avoid scientific notation
    df["music_id"] = df["music_id"].astype("Int64").astype(str)
    df["id"] = df["id"].astype("Int64").astype(str)
    df["playlist_id"] = df["playlist_id"].astype("Int64").astype(str)
    
    df["hashtag_names"] = df["hashtag_names"].apply(lambda x: ", ".join(x) if isinstance(x, list) else x)

    return df
```

```python
from deep_translator import GoogleTranslator

"""
Detects the language of the input text and translates it to English if it's not already in English.
If the text is empty or not a string, it returns the original text.
If an error occurs during translation, it prints the error and returns the original text.
"""
def detect_translate(text):
    try:
        if not isinstance(text, str) or text.strip() == "":
            return text  # skips empty or non-strings

        return GoogleTranslator(source='auto', target='en').translate(text)
    
    except Exception as e:
        print(f"Error translating: {text[:30]}... // {e}")
        return text
```

```python
import re
from collections import Counter

brand_list = [
    # U.S. & Global
    "L'Oréal", "Maybelline", "Revlon", "CoverGirl", "Estée Lauder", "Clinique",
    "MAC", "Smashbox", "Too Faced", "Urban Decay", "NARS", "Benefit",
    "Anastasia Beverly Hills", "Fenty Beauty", "Rare Beauty", "Glossier",
    "Charlotte Tilbury", "Pat McGrath", "Tarte", "KVD Beauty", "E.l.f.",
    "Morphe", "ColourPop", "NYX", "Milk Makeup", "Kylie Cosmetics", "KKW Beauty",
    "Hourglass", "Bobbi Brown", "Laura Mercier", "Shiseido", "IT Cosmetics",
    "Huda Beauty", "KKW", "Sephora Collection", "Sigma Beauty", "P. Louise",
    "Drunk Elephant", "Rhode", "Makeup by Mario",

    # Korean (K-Beauty)
    "Etude House", "Innisfree", "Laneige", "Sulwhasoo", "Missha", "Nature Republic",
    "The Face Shop", "TonyMoly", "Holika Holika", "Banila Co", "Dr.Jart+",
    "Cosrx", "Mediheal", "Clio", "Peripera", "Hera", "Amorepacific",
    "Rom&nd", "Beauty of Joseon", "Pyunkang Yul", "Iope", "Skinfood", "Medicube",
    "Biodance", "Anua",

    # Japanese (J-Beauty)
    "Shiseido", "SK-II", "DHC", "Canmake", "Kose", "Majolica Majorca",
    "Muji Skincare", "Hada Labo", "Sana", "Cezanne", "Kate Tokyo",
    "Suqqu", "PolA", "Albion", "MAQuillAGE",

    # Chinese
    "Perfect Diary", "Florasis", "Chando", "Marie Dalgar", "Peach C", "Herborist",
    "Winona", "Judydoll", "Carslan", "Into You", "Catkin", "ZEESEA", "SACE LADY",

    # European Luxury
    "Chanel", "Dior", "Yves Saint Laurent", "Guerlain", "Givenchy", "Lancôme",
    "Sisley", "By Terry", "Clarins", "La Prairie", "Biotherm", "Filorga", "YSL",

    # UK & EU Indie / Popular
    "Revolution Beauty", "Illamasqua", "No7", "Soap & Glory", "Catrice", "Essence",
    "Kiko Milano", "Sleek Makeup", "Barry M",

    # Indian & South Asia
    "Lakme", "Lotus Herbals", "Forest Essentials", "Biotique", "Mamaearth",
    "Sugar Cosmetics", "Nykaa Cosmetics", "Kay Beauty", "Colorbar",

    # Middle East
    "Huda Beauty", "K7L", "The Camel Soap Factory", "Glambeaute",
    "Mikyajy", "Wow Beauty",

    # Australia & NZ
    "Napoleon Perdis", "Australis", "Go-To Skincare", "Frank Body", "Sukin",
    "Antipodes", "Aesop",

    # Latin America
    "Natura", "Boticário", "Quem Disse Berenice", "Vult", "Eudora",
    "Lola Cosmetics", "Ésika", "Cyzone", "Belcorp",

    # Skincare-Specific Giants
    "Cetaphil", "Neutrogena", "Aveeno", "Olay", "CeraVe", "La Roche-Posay",
    "Vichy", "Eucerin", "Bioderma", "Kiehl's", "Origins", "Philosophy", 
    "Tatcha", "Youth to the People", "The Ordinary",
    "Paula's Choice", "Sunday Riley", "Murad", "First Aid Beauty",
    "Glow Recipe", "Versed", "Dermalogica", "Obagi", "Peter Thomas Roth",
    "Fresh", "Sisley Paris", "Made By Mitchell", "Garnier", "SHEGLAM",
    
    # Wig?
    "Chelsea Smith Crowns"
]

def build_brand_patterns(brand_list):
    """Return list of (canonical_brand, regex_patterns)."""
    patterns = []
    seen = set()
    for b in brand_list:
        if not b or not isinstance(b, str):
            continue
        canon = b.strip()
        key = canon.lower()
        if key in seen:
            continue
        seen.add(key)

        # split into alphanumeric chunks to keep unicode letters
        parts = re.findall(r'\w+', canon, flags=re.UNICODE)
        if not parts:
            continue

        # allows optional whitespace between parts:
        # e.g. "Rare Beauty" -> r"\brare\s*beauty\b"
        pat_allow_spaces = re.compile(
            r'\b' + r'\s*'.join(re.escape(p) for p in parts) + r'\b',
            re.IGNORECASE
        )

        # matches concatenated alphanumeric form:
        # e.g. "Rare Beauty" -> r"\brarebeauty\b"
        concat = ''.join(parts)
        pat_concat = re.compile(r'\b' + re.escape(concat) + r'\b', re.IGNORECASE)

        # matches brand as a prefix in longer tokens:
        # e.g. "nyxcosmetics" -> "NYX"
        pat_prefix = re.compile(r'\b' + re.escape(concat) + r'\w*\b', re.IGNORECASE)

        patterns.append((canon, [pat_allow_spaces, pat_concat, pat_prefix]))
    return patterns


def filter_brand_counts(word_counts, brand_patterns):
    """
    Given a Counter of tokens (word_counts) and brand_patterns list,
    return (Counter aggregated to canonical brands, dict mapping original_token->matched_brand).
    """
    out = Counter()
    token_map = {}

    for token, cnt in word_counts.items():
        tok = str(token)
        for canon, regexes in brand_patterns:
            if any(r.search(tok) for r in regexes):
                out[canon] += cnt
                token_map[tok] = canon
                break
    return out, token_map
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def create_graph(filtered_counts):
    """
    Create a bar plot of the top 50 most frequently mentioned brands.
    """
    # Get top 50 most common words
    top_brands = filtered_counts.most_common(50)
    words, counts = zip(*top_brands)

    # Plot using seaborn
    plt.figure(figsize=(12, 8))
    sns.barplot(x=list(counts), y=list(words), palette="viridis")

    plt.title("Top 50 Most Frequently Mentioned Brands", fontsize=16)
    plt.xlabel("Frequency", fontsize=14)
    plt.ylabel("Brands", fontsize=14)
    plt.show()
```
